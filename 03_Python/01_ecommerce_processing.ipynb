```python
df = df_order_items

cols_dates = ['created_at', 'shipped_at', 'delivered_at', 'returned_at']

# Parser les dates
df_order_items[cols_dates] = df_order_items[cols_dates].apply(lambda x: pd.to_datetime(x, errors='coerce'))
```

```python
df_order_items["order_preparation_time"] = round((df_order_items["shipped_at"] - df_order_items["created_at"] )/ pd.Timedelta(hours=1), 0)
df_order_items['order_preparation_time'] = df_order_items['order_preparation_time'].fillna(0)
df_order_items['order_preparation_time'] = df_order_items['order_preparation_time'].astype(int)

df_order_items["order_delivery_time"] = round((df_order_items["delivered_at"] - df_order_items["created_at"] )/ pd.Timedelta(hours=1), 0)
df_order_items['order_delivery_time'] = df_order_items['order_delivery_time'].fillna(0)
df_order_items['order_delivery_time'] = df_order_items['order_delivery_time'].astype(int)

df_order_items.head()
```

```python
df_order_items= df_order_items[~df_order_items['order_preparation_time'] < 0]
```

#### â—» Orders Dataset

```python
# Note that we are not using the orders table here since informations are already in the order_item table

# --- Merging datasets into one
orders = df_order_items.merge(df_products, how='left', on=['product_id']) \
                   .merge(df_centers, how='left', on=['distribution_center_id']) \
                   .merge(df_inventory, how='left', on=['inventory_item_id']) \
                   .merge(df_users, how='left', on=['user_id'])

# --- Display
df = orders.copy()
df.head()
```

```python
df['profit'] = df['order_sale_price'] - df['product_cost']
```
